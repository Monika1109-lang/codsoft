```python
from sklearn.linear_model import LogisticRegression
```

```python
from imblearn.over_sampling import SMOTE
data = df.dropna(subset=['Class'])  # Drop rows where 'Class' is NaN
X = df.drop('Class', axis=1)
y = df['Class']
y = y.fillna(y.mode()[0])  # Replace NaN with the most frequent value

```

```python
# Check for missing values in y
missing_values_count = y.isnull().sum()  # Calculate the number of NaN values
print(f"Missing values in y: {missing_values_count}")

# Handle missing values if any exist
if missing_values_count > 0:
    print("Handling missing values in y...")
    y = y.fillna(y.mode()[0])  # Replace NaN values with the most frequent value

```

```python
print(f"Missing values in X before preprocessing: {X.isnull().sum().sum()}")

```

```python
X = X.fillna(X.median())  # Replace NaN values with the median
data = data.dropna()  # Drops rows with NaN in any column
X = data.drop('Class', axis=1)
y = data['Class']
```

```python
# Normalize the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Apply SMOTE
smote = SMOTE(random_state=42)
X_resampled, y_resampled = smote.fit_resample(X_scaled, y)

# Check shapes
print(f"Original shape: X={X.shape}, y={y.shape}")
print(f"Resampled shape: X={X_resampled.shape}, y={y_resampled.shape}")
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X_resampled, y_resampled, test_size=0.2, random_state=42
)


log_reg = LogisticRegression(random_state=42)
log_reg.fit(X_train, y_train)


```

```python

from sklearn.metrics import classification_report, accuracy_score

X_train, X_test, y_train, y_test = train_test_split(X_resampled, y_resampled, test_size=0.3, random_state=42)


rf_classifier = RandomForestClassifier(n_estimators=100, random_state=42)
rf_classifier.fit(X_train, y_train)


y_pred = rf_classifier.predict(X_test)


print("Classification Report:")
print(classification_report(y_test, y_pred))

print(f"Accuracy: {accuracy_score(y_test, y_pred):.2f}")
```

```python

from sklearn.metrics import classification_report, precision_score, recall_score, f1_score
print(f"Precision: {precision_score(y_test, y_pred):.2f}")
print(f"Recall: {recall_score(y_test, y_pred):.2f}")
print(f"F1-Score: {f1_score(y_test, y_pred):.2f}")
```
